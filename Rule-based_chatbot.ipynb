```python
import pandas as pd
import random
```

```python
# Load the dataset
dataset = pd.read_csv("cafedataset.csv")
# Remove NaN values from the DataFrame
dataset.dropna(subset=["Question", "answer"], inplace=True)
```

```python
# Create a dictionary of rules based on the dataset
rules = {}
for i in range(len(dataset)):
    question = dataset.iloc[i]['Question']
    answer = dataset.iloc[i]['answer']
    if question not in rules:
        rules[question] = [answer]
    else:
        rules[question].append(answer)
```

```python
# Define a function to handle user input and return a response
def chatbot_response(user_input):
    for key in rules:
        if key in user_input.lower():
            return random.choice(rules[key])
    return "I'm sorry, I don't understand your question."
```

```python
# Test the chatbot
correct_responses = 0
total_responses = 0
for i in range(len(dataset)):
    user_input = dataset.iloc[i]['Question']
    expected_response = dataset.iloc[i]['answer']
    chatbot_response_val = chatbot_response(user_input)
    total_responses += 1
    if chatbot_response_val == expected_response:
        correct_responses += 1
    print("User: " + user_input)
    print("Chatbot: " + str(chatbot_response_val))
    print("Expected response: " + str(expected_response))
    print("")
```

```python
# Calculate the accuracy
accuracy = (correct_responses / total_responses) * 100
print("Accuracy: {:.2f}%".format(accuracy))
```

```python
# Run the chatbot
while True:
    user_input = input("user: ")
    response = chatbot_response(user_input)
    print("Cafe-Chatbot: " + response)
```
